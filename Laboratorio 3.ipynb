# Ejercicio 8
### Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
# Seleccionamos los datos de Madrid
madrid = df.loc[df["Municipio"] == "Madrid"].iloc[0]
madrid_lat = np.radians(madrid["Y"])
madrid_lon = np.radians(madrid["X"])

# Función para calcular la distancia entre dos puntos geográficos
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # radio de la Tierra en km
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    return R * c

# Calculamos la distancia entre cada municipio y Madrid
distances = []
for i, row in df.iterrows():
    if row["Municipio"] == "Madrid":
        continue
    lat = np.radians(row["LAT"])
    lon = np.radians(row["LON"])
    d = haversine(madrid_lat, madrid_lon, lat, lon)
    distances.append((row["Municipio"], d))

# Encontramos el municipio más cercano y el más lejano a Madrid
closest = min(distances, key=lambda x: x[1])
farthest = max(distances, key=lambda x: x[1])

# Mostramos los resultados
print("El municipio más cercano a Madrid es:", closest[0])
print("El municipio más lejano a Madrid es:", farthest[0])
```
